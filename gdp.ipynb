```python
import pandas as pd
import panel as pn
from pathlib import Path
import hvplot.pandas
```

```python
real_gdp_path = Path("data/real_gdp.csv")
real_gdp      = pd.read_csv(real_gdp_path, index_col='DATE', parse_dates=True, infer_datetime_format=True)
real_gdp      = real_gdp.rename(columns={"GDPC1": "Billions of Dollars"})

real_gdp_per_capita_path = Path("data/real_gdp_per_capita.csv")
real_gdp_per_capita      = pd.read_csv(real_gdp_per_capita_path,index_col='DATE', parse_dates=True, infer_datetime_format=True)
real_gdp_per_capita      = real_gdp_per_capita.rename(columns={"A939RX0Q048SBEA": "Thousands of Dollars"})

inflation_path = Path("data/inflation.csv")
inflation      = pd.read_csv(inflation_path, index_col='date', parse_dates=True, infer_datetime_format=True)
```

```python
real_gdp_all      = real_gdp.iloc[::-1]
real_gdp_all      = real_gdp_all.iloc[0:160:4]
real_gdp_all_plot = real_gdp_all.iloc[0:160:4].hvplot.line(title="Real GDP 1981 - 2020", xlabel="Date")

real_gdp_per_capita_all = real_gdp_per_capita[::-1]
real_gdp_per_capita_all = real_gdp_per_capita_all.iloc[0:160:4]
real_gdp_per_capita_all_plot = real_gdp_per_capita_all.iloc[0:160:4].hvplot.line(title="Real GDP per Capita 1981 - 2020", xlabel="Date")

inflation_all      = inflation.iloc[::-1]
inflation_all      = inflation_all.iloc[0:40]
inflation_all_plot = inflation_all.iloc[0:40].hvplot.line(title="Inflation 1981 - 2020", ylabel="Percent", xlabel="Date")

all_column =  pn.Column(real_gdp_all_plot, real_gdp_per_capita_all_plot, inflation_all_plot)
```

```python
reagan_rgdp      = real_gdp_all.iloc[32:40].hvplot.line(title="Real GDP 1981 - 1988")
reagan_rgdpc     = real_gdp_per_capita_all.iloc[31:80].hvplot.line(title="Real GDP per Capita 1981 - 1988")
reagan_inflation = inflation_all.iloc[31:80].hvplot.line(title="Inflation 1981 - 2020")

reagan_column =  pn.Column(reagan_rgdp, reagan_rgdpc, reagan_inflation)
```

```python
bush_sr_rgdp       = real_gdp_all.iloc[28:32].hvplot.line(title="Real GDP 1989 - 1992")
bush_sr_rgdpc      = real_gdp_per_capita_all.iloc[28:32].hvplot.line(title="Real GDP per Capita 1989 - 1992")
bush_sr_inflation  = inflation_all.iloc[28:32].hvplot.line(title="Inflation 1989 - 1992")

bush_sr_column =  pn.Column(bush_sr_rgdp, bush_sr_rgdpc, bush_sr_inflation)
```

```python
clinton_rgdp       = real_gdp_all.iloc[20:28].hvplot.line(title="Real GDP 1993 - 2000")
clinton_rgdpc      = real_gdp_per_capita_all.iloc[20:28].hvplot.line(title="Real GDP per Capita 1993 - 2000")
clinton_inflation  = inflation_all.iloc[20:28].hvplot.line(title="Inflation 1993 - 2000")

clinton_column =  pn.Column(clinton_rgdp, clinton_rgdpc, clinton_inflation)
```

```python
bush_jr_rgdp       = real_gdp_all.iloc[12:20].hvplot.line(title="Real GDP 2001 - 2008")
bush_jr_rgdpc      = real_gdp_per_capita_all.iloc[12:20].hvplot.line(title="Real GDP per Capita 2001 - 2008")
bush_jr_inflation  = inflation_all.iloc[12:20].hvplot.line(title="Inflation 2001 - 2008")

bush_jr_column =  pn.Column(bush_jr_rgdp, bush_jr_rgdpc, bush_jr_inflation)
```

```python
obama_rgdp       = real_gdp_all.iloc[4:12].hvplot.line(title="Real GDP 2009 - 2016")
obama_rgdpc      = real_gdp_per_capita_all.iloc[4:12].hvplot.line(title="Real GDP per Capita 2008 - 2016")
obama_inflation  = inflation_all.iloc[4:12].hvplot.line(title="Inflation 2009 - 2016")

obama_column =  pn.Column(obama_rgdp, obama_rgdpc, obama_inflation)
```

```python
trump_rgdp       = real_gdp_all.iloc[0:4].hvplot.line(title="Real GDP 2017 - 2020")
trump_rgdpc      = real_gdp_per_capita_all.iloc[0:4].hvplot.line(title="Real GDP per Capita 2017 - 2020")
trump_inflation  = inflation_all.iloc[0:4].hvplot.line(title="Inflation 2017 - 2020")

trump_column =  pn.Column(trump_rgdp, trump_rgdpc, trump_inflation)
```

```python
dashboard = pn.Tabs(
    ("1981 - 2020", all_column),
    ("Reagan", reagan_column),
    ("Bush Sr", bush_sr_column),
    ("Clinton", clinton_column),
    ("Bush Jr", bush_jr_column),
    ("Obama", obama_column),
    ("Trump", trump_column)
)

dashboard.servable()
```
